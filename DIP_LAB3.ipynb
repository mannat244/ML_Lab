```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Image size
h, w = 256, 256

img = np.ones((h, w, 3), dtype=np.uint8) * 180

cv2.rectangle(img, (40, 40), (120, 120), (0, 0, 0), -1)

cv2.circle(img, (180, 80), 30, (255, 255, 255), -1)

cv2.line(img, (0, 200), (256, 200), (100, 100, 100), 2)

# Save image
cv2.imwrite("synthetic.png", img)

# Display image
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title("Synthetic Image")
plt.axis("off")
plt.show()
```

# Histogram equalisation

```python
# ================== IMPORTS ==================
import cv2
import numpy as np
import matplotlib.pyplot as plt

# ================== LOAD IMAGE ==================
# Upload op.png before running this cell
img = cv2.imread("op.png", cv2.IMREAD_GRAYSCALE)

assert img is not None, "op.png NOT loaded. Please upload the image."

# ================== HISTOGRAM EQUALIZATION ==================
equalized = cv2.equalizeHist(img)

# ================== DISPLAY RESULTS (BLOCK) ==================
plt.figure(figsize=(12, 4))

plt.subplot(1, 3, 1)
plt.title("Original Image (op.png)")
plt.imshow(img, cmap="gray")
plt.axis("off")

plt.subplot(1, 3, 2)
plt.title("Histogram Equalized Image")
plt.imshow(equalized, cmap="gray")
plt.axis("off")

plt.subplot(1, 3, 3)
plt.title("Histogram Comparison")
plt.hist(img.ravel(), bins=256, range=[0, 256], alpha=0.6, label="Original")
plt.hist(equalized.ravel(), bins=256, range=[0, 256], alpha=0.6, label="Equalized")
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt


A = np.zeros((1024, 1024), dtype=np.uint8)
for i in range(0, 1024, 256):
    A[i:i+128, :] = 255
    A[i+128:i+256, :] = 0


B = np.zeros((1024, 1024), dtype=np.uint8)
for j in range(0, 1024, 256):
    B[:, j:j+128] = 255
    B[:, j+128:j+256] = 0


add_img = np.clip(A + B, 0, 255)
sub_img = np.clip(A.astype(int) - B.astype(int), 0, 255).astype(np.uint8)
mul_img = (A * B) // 255
avg_img = ((A + B) // 2).astype(np.uint8)


plt.figure(figsize=(12, 10))

plt.subplot(2, 3, 1)
plt.imshow(A, cmap='gray')
plt.title("Image A (Horizontal)")
plt.axis('off')

plt.subplot(2, 3, 2)
plt.imshow(B, cmap='gray')
plt.title("Image B (Vertical)")
plt.axis('off')

plt.subplot(2, 3, 3)
plt.imshow(add_img, cmap='gray')
plt.title("A + B")
plt.axis('off')

plt.subplot(2, 3, 4)
plt.imshow(sub_img, cmap='gray')
plt.title("A - B")
plt.axis('off')

plt.subplot(2, 3, 5)
plt.imshow(mul_img, cmap='gray')
plt.title("A  B")
plt.axis('off')

plt.subplot(2, 3, 6)
plt.imshow(avg_img, cmap='gray')
plt.title("(A + B) / 2")
plt.axis('off')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt


rows = 256
cols = 1024

x = np.linspace(0, 2 * np.pi, cols)


sin_wave = (np.sin(x) + 1) / 2 * 255

sinusoidal_img = np.tile(sin_wave, (rows, 1)).astype(np.uint8)

plt.imshow(sinusoidal_img, cmap='gray')
plt.title("256x1024 Sinusoidal Grayscale Image")
plt.axis('off')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

img = np.ones((256, 256), dtype=np.uint8) * 255   # white image

box_size = 58

start = (256 - box_size) // 2
end = start + box_size

img[start:end, start:end] = 0

plt.imshow(img, cmap='gray')
plt.title("256x256 White Image with Centered Black Box (58x58)")
plt.axis('off')
plt.show()
```
